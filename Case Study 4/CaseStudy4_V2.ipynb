```python
dfx = df1.loc[df1['TOTAL SPECIMENS'] != 'X']
```

```python
dfx
```

```python
dfx['TOTAL SPECIMENS'] = dfx['TOTAL SPECIMENS'].astype(np.float64)
```

```python
from datetime import datetime
import pandas as pd

def year_week(y, w):
    return datetime.strptime(f'{y} {w} 1', '%G %V %u')


dfx['year_week_ts'] = dfx.apply(lambda row: year_week(row.YEAR, row.	WEEK), axis=1)
dfx['year_week_str'] = dfx.apply(lambda row: row.year_week_ts.strftime('%G-%V'), axis=1)

dfx.head(32)
```

```python
dfx.info()
```

```python
import seaborn as sns
import numpy as np
#import altair as alt
import matplotlib.pyplot as plt
# display boxplot 
bx = dfx.boxplot(column = 'TOTAL SPECIMENS', by = ['YEAR', 'WEEK'], figsize=(14,7))
plt.xticks(rotation=45)
bx
```

```python
# plot
sns.set_style('ticks')
fig, ax = plt.subplots()
# the size of A4 paper
fig.set_size_inches(11.7, 8.27)
sns.violinplot(x= 'WEEK', y='TOTAL SPECIMENS', data=dfx)    
sns.despine()
plt.title('Violin Plot Specimens by Year')
plt.xticks(rotation=45)
```

```python
air_pop=dfx['TOTAL SPECIMENS']
air_pop.index=dfx['WEEK']
print(air_pop.head())
fig = plt.figure()

air_pop.plot()
plt.title("A")
plt.xlabel("T")
plt.ylabel('P')
plt.show()
```
