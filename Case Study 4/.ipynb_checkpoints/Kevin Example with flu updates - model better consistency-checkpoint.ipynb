## Introduction/Overview

Time series analysis comprises methods for analyzing time series data in order to extract meaningful statistics and other characteristics of the data. Time series forecasting is the use of a model to predict future values based on previously observed values.
Time series are widely used for non-stationary data, like economic, weather, stock price, and retail sales in this post. We will demonstrate different approaches for forecasting retail sales time series. Let’s get started

Time Series
- time dependent, observations are not independent (which is assumption of linear regression, so need to consider other types of analysis.)
- likely has an increasing or decreasing trend, overall
- most have seasonality trends,

Important aspects of Time Series data, what are the goals, what are we trying to achieve as part of analysis

looking for Stationarity - statistical properities of mean, variance remain constant overtime; also autocovariance does not depend on time.
assumption is foundation for most time series models

exhibit behavior over time, high likelihood to continue into future

visual analysis: what does it tell us:
- seasonaly peaks
- overall trend of entire plot

Formal checks for Stationarity:
- plot rolling stats, plot moving average, moving variance, both to see relationship with time, visual technique,  
- Dickey-Fuller Test, null hypothesis is that TS is not stationary, consider test statistic and critical value, note - consider signed values (not absolute)


### How to make Time Series Stationary
make as close as possible
trend - varying mean over time
seasonality - variations at specific time frames, 
goal is to identify and remove

**Trend**
-transformation - penalize part of trend, i.e. higher or lower values, long, square root, cube root, log transform
- aggregation - take average from time period 
- smoothing - take rolling average, rquires time period to be strictly defined, can do a weighted moving average, exponentially weighted moving average
- polynomial fitting - fit regression model


**Seasonality**


Ways to Remove
- differencing - take difference with particular time lag, deals with both trend and seasonality, take difference of observation at particular instant compared to previous instant, improves stationarity, measured in "orders" - first, second, third. then again compare test statistic and critical value as part of Dickey-Fuller test
- decompostion - model trend and seasonality and remove from model, model both separately, returns trend, seasonality, residuals, then compare test statistics and critical vaue as part of Dickey-Fuller test

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries, second):
    
    #Determing rolling statistics
    rolmean = second.rolling(52).mean()
    rolstd = second.rolling(52).std()

    #Plot rolling statistics:
    plt.figure(figsize=(20, 4))
    orig = plt.plot(second, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
from statsmodels.tsa.stattools import acf, pacf
acf_dif_y = acf(y.dropna(), nlags=20)
#Plot ACF: 
plt.figure(figsize=(20, 4))

plt.subplot(121) 
plt.plot(acf_dif_y)
plt.xticks(np.arange(21))
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(acf_dif_y)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(acf_dif_y)),linestyle='--',color='gray')
for i in range(1,20):
    plt.axvline(x=i,linestyle=':',color='gray')
plt.title('Autocorrelation Function')
```

There appears to be serial auto correlation. Suggestion that the data may need to be differenced. We will run the dicky fuller test to check this. 

Dickey-Fuller Test - looks to see if the rolling average increases or decreases, which is a measure of stationarity
(lag number of values used to compute the rolling statistics)

The null hypothesis of the Augmented Dickey-Fuller is that there is a unit root(it is non-stationary), with the alternative that there is no unit root. If the pvalue is above a critical size, then we cannot reject that there is a unit root.

The p-values are obtained through regression surface approximation from MacKinnon 1994, but using the updated 2010 tables. If the p-value is close to significant, then the critical values should be used to judge whether to reject the null.

If there is a unit root, there is evidence of non-stationarity data

The ACF stands for Autocorrelation function, and the PACF for Partial Autocorrelation function. Looking at these two plots together can help us form an idea of what models to fit. Autocorrelation computes and plots the autocorrelations of a time series. Autocorrelation is the correlation between observations of a time series separated by k time units.
Similarly, partial autocorrelations measure the strength of relationship with other terms being accounted for, in this case other terms being the intervening lags present in the model. For example, you can see a large spike at lag_17 and Lag_16 for the transformed data and then again at the 52. This makes scene as you would expect the same time the previous year would have a similar count of cases. 
 

Autocorrelation Function (ACF) - measure of correlation between time series and a lagged version of itself

Partial Autocorrelation Function (PACF) - measures correlation between time series and lagged version after eliminating variation already explained by intervening comparisons (??)

# Time series forecasting with ARIMA
We are going to apply one of the most commonly used method for time-series forecasting, known as ARIMA, which stands for Autoregressive Integrated Moving Average.
ARIMA models are denoted with the notation ARIMA(p, d, q). These three parameters account for seasonality, trend, and noise in data:

AutoRegressive - current prediction depends on past results, lags of dependent variable
Integrated - differences between past results are important, lagged forecast errors in predicted equation, 
Moving Average - average of previous results are important, nonseasonal differences

p - autoregressive part
d - integrated
q - moving average

ideal scenario with time series - no trend, no seasonality
**ONCE STATIONARY, can proceed with analysis

Important Rules: which ones to include?

**Can set p with lag value of PACF where crosses upper confidence interval for first time
**Can set q with lag value of ACT where cross upper confidence interval for first time

ARIMA Forecasting Method for Time Series

# Validating forecasts
To help us understand the accuracy of our forecasts, we compare predicted values to real values of the time series, and we set forecasts to start at 2017–01–01 to the end of the data.

```python
pred = results.get_prediction(start=150, dynamic=False)
pred_ci = pred.conf_int()
ax = dfx3.dropna().plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('Positive Cases')
plt.legend()
plt.show()
```

The line plot is showing the observed values compared to the rolling forecast predictions. Overall, our forecasts align with the true values very well, showing an upward trend starts from the beginning of the year and captured the seasonality toward the end of the year.
